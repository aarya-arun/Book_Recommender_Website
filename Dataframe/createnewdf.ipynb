```python
import pandas as pd
import numpy as np
import matplotlib as plt
import re

%matplotlib inline

df=pd.read_csv("Downloads/goodbooks/books.csv")
df.head()
```

```python
to_drop=['id','book_id','best_book_id','work_id','books_count','isbn13','title','work_ratings_count','ratings_1','ratings_2','ratings_3','ratings_4','ratings_5']
df.drop(to_drop, inplace=True, axis=1)
```

```python
df
```

```python
df=df.dropna()
df.reset_index(drop=True, inplace=True)
```

```python
df
```

```python
#Change publication year to integer 
df['original_publication_year'] = df['original_publication_year'].astype(int)
```

```python
df
```

```python
for i in range(len(df)):
     df['authors'][i] = df.authors[i].split(",")
        
```

```python
for i in range(len(df)):
    df['authors'][i]=df['authors'][i][0]
```

```python
df
```

```python
regex = re.compile('[^a-zA-Z]')
authorsmatch=[]


for i in range(len(df)):
    authorsmatch.append(regex.sub('',df['authors'][i]))
    
df['authorsmatch']=authorsmatch
```

```python
df.head()
```

```python
for i in range(len(df)):
    df['authorsmatch'][i]=df['authorsmatch'][i].lower()
```

```python
df.head()
```

```python
regex = re.compile('[^a-zA-Z]')
titlesmatch=[]


for i in range(len(df)):
    titlesmatch.append(regex.sub('',df['original_title'][i]))
    
df['titlesmatch']=titlesmatch


for i in range(len(df)):
    df['titlesmatch'][i]=df['titlesmatch'][i].lower()
```

```python
df.head()
```
